```python
from backend.database.mongodb import get_jobs_collection
from backend.utils.job_search import JobListing  # Import our Pydantic model
from pprint import pprint

# Get the jobs collection
jobs_collection = get_jobs_collection()

# Find AI Engineer positions
ai_engineer_jobs = list(
    jobs_collection.find({
        "title": {"$regex": "AI Engineer", "$options": "i"}
    })
    .limit(10)
)
```

```python
# Convert MongoDB documents to Pydantic models and print
for job_doc in ai_engineer_jobs:
    # Convert MongoDB document to Pydantic model
    job = JobListing(**job_doc)
    
    print("\n" + "="*100)
    
    # Use Pydantic's model_dump() for pretty printing
    pprint(job.model_dump(exclude={'extensions'}), 
           width=100, 
           sort_dicts=False)
    
    # Print extensions separately if they exist
    if job.detected_extensions:
        print("\nExtensions:")
        pprint(job.detected_extensions.model_dump(exclude_none=True),
               width=100)

print(f"\nTotal AI Engineer positions found: {len(ai_engineer_jobs)}")
```

```python
from backend.database.mongodb import get_jobs_collection
from backend.utils.job_search import JobListing
import json
from datetime import datetime
from pathlib import Path

# Get the jobs collection
jobs_collection = get_jobs_collection()

# Find AI Engineer positions
ai_engineer_jobs = list(
    jobs_collection.find({
        "title": {"$regex": "AI Engineer", "$options": "i"}
    })
    .limit(10)
)

# Convert to Pydantic models and then to dict
jobs_data = [JobListing(**job).model_dump() for job in ai_engineer_jobs]

# Create a timestamp for the filename
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = Path(f"ai_engineer_jobs_{timestamp}.json")

# Save to JSON file with nice formatting
with output_file.open('w', encoding='utf-8') as f:
    json.dump(
        jobs_data, 
        f, 
        indent=2, 
        default=str,  # Handles datetime serialization
        ensure_ascii=False  # Properly handles unicode characters
    )

print(f"Saved {len(jobs_data)} jobs to {output_file}")
```

```python
from backend.database.mongodb import get_jobs_collection
from backend.utils.job_search import JobListing
from datetime import datetime
from pathlib import Path
from pprint import pformat

# Get the jobs collection
jobs_collection = get_jobs_collection()

# Find AI Engineer positions
ai_engineer_jobs = list(
    jobs_collection.find({
        "title": {"$regex": "Software Engineer", "$options": "i"}
    })
    .limit(20)
)

# Create a timestamp for the filename
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = Path(f"ai_engineer_jobs_{timestamp}.txt")

# Save to text file with formatted output
with output_file.open('w', encoding='utf-8') as f:
    for i, job_dict in enumerate(ai_engineer_jobs, 1):
        # Convert MongoDB document to Pydantic model
        job = JobListing(**job_dict)
        
        f.write(f"\n{'='*100}\n")
        f.write(f"Job {i} of {len(ai_engineer_jobs)}\n")
        f.write(f"{'='*100}\n\n")
        
        # Write key information
        f.write(f"Title: {job.title}\n")
        f.write(f"Company: {job.company_name}\n")
        f.write(f"Location: {job.location}\n\n")
        
        # Write job description
        f.write("Description:\n")
        f.write(f"{job.description}\n\n")
        
        # Write job highlights
        f.write("Highlights:\n")
        for highlight in job.job_highlights:
            f.write(f"\n{highlight.title}:\n")
            for item in highlight.items:
                f.write(f"- {item}\n")
        
        # Write extensions if they exist
        if job.detected_extensions:
            f.write("\nAdditional Information:\n")
            ext = job.detected_extensions
            if ext.posted_at:
                f.write(f"Posted: {ext.posted_at}\n")
            if ext.salary:
                f.write(f"Salary: {ext.salary}\n")
            if ext.schedule:
                f.write(f"Schedule: {ext.schedule}\n")
        
        f.write(f"\nApply Link: {job.apply_link}\n\n")

print(f"Saved {len(ai_engineer_jobs)} jobs to {output_file}")
```
